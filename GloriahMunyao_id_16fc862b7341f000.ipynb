```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error

# Loading the dataset
file_path = r'C:\Hamoye\energydata_complete.csv'
data = pd.read_csv(file_path)

# Removing columns "date" and "lights"
removeColumns = ["date", "lights"]
data = data.drop(removeColumns, axis=1)

# Normalizing the dataset using MinMaxScaler
scaler = MinMaxScaler()
normalizedData = scaler.fit_transform(data)

# Converting the normalized data back to a DataFrame
normalizedData = pd.DataFrame(normalizedData, columns=data.columns)

#Target variable as "Appliances"
target = normalizedData["Appliances"]

#Features variable to be all columns except "Appliances"
features = normalizedData.drop("Appliances", axis=1)


# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    features, target, test_size=0.3, random_state=42
)

# Train a Ridge regression model with alpha=0.4
ridgeModel = Ridge(alpha=0.4)
ridgeModel.fit(X_train, y_train)

# Make predictions on the test set using the Ridge model
temp2_ridge = ridgeModel.predict(X_test)

# Calculate the RMSE for the Ridge model
RMSEridge = mean_squared_error(y_test, temp2_ridge, squared=False)

print("RMSE with Ridge Regression:", RMSEridge)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import Lasso

# Load the dataset
file_path = r'C:\Hamoye\energydata_complete.csv'
data = pd.read_csv(file_path)

# Remove columns "date" and "lights"
columns_to_remove = ["date", "lights"]
data = data.drop(columns_to_remove, axis=1)

# Normalize the dataset using MinMaxScaler
scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(data)

# Convert the normalized data back to a DataFrame
normalized_data = pd.DataFrame(normalized_data, columns=data.columns)

# Set the target variable as "Appliances"
target = normalized_data["Appliances"]

# Set the features (all columns except "Appliances")
features = normalized_data.drop("Appliances", axis=1)

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)

# Fit a Lasso regression model
lasso_model = Lasso(alpha=0.001)
lasso_model.fit(X_train, y_train)

# Get the feature weights
feature_weights = pd.Series(lasso_model.coef_, index=features.columns)

# Count the number of features with non-zero weights
non_zero_features = sum(feature_weights != 0)

print("Number of features with non-zero feature weights:", non_zero_features)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error

# Use the Lasso regression model to make predictions on the test set
lasso_predictions = lasso_model.predict(X_test)

# Calculate the RMSE
lasso_rmse = np.sqrt(mean_squared_error(y_test, lasso_predictions))

# Print the RMSE with three decimal places
print("RMSE with Lasso regression:", '{:.3f}'.format(lasso_rmse))
```
